```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
```

```python
def collatz(n):
    """Collatz conjecture.
    
    Calculate the collatz sequence given a starting
    number `n` and return the values as a list of integers.
    
    Parameters
    ----------
    n : int
        starting number
        
    Returns
    -------
    l : list of ints
    
    Examples
    --------
    >>> collatz(7)
    [7, 22, 11, 34, 17, 52, 26, 13, 40, 20, 10, 5, 16, 8, 4, 2, 1]
    """
    if n <= 0:
        return []
    res = [n]
    while n != 1:
        n = n / 2 if n % 2 == 0 else n*3 + 1
        res.append(n)
    return res        
```
